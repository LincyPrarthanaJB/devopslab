```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
model="gemini-2.0-flash", # or "gemini-2.5-pro" etc.
temperature=0.7, google_api_key=os.environ["api_key"]
)
```

```python
import tkinter as tk
from tkinter import ttk
from langchain.prompts import FewShotPromptTemplate, PromptTemplate
from langchain.chains import LLMChain


# Examples
examples = [
    {"word": "happy", "antonym": "sad"},
    {"word": "increase", "antonym": "decrease"}
]

example_prompt = PromptTemplate(
    input_variables=["word", "antonym"],
    template="Word: {word} â†’ Antonym: {antonym}"
)

prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    prefix="Provide antonyms:",
    suffix="Word: {input}\nAntonym:",
    input_variables=["input"]
)

# Create the LLMChain.
if llm:
    chain = LLMChain(llm=llm, prompt=prompt)
else:
    chain = None

def find_antonym():
    """Fetches input from the GUI, runs the LangChain process, and displays the output."""
    if not chain:
        output_text.delete(1.0, tk.END)
        output_text.insert(tk.END, "Error: Language model not configured. Please edit the script to add your LLM.")
        return

    input_word = input_entry.get()
    if input_word.strip() == "":
        output_text.delete(1.0, tk.END)
        output_text.insert(tk.END, "Please enter a word.")
        return

    # Run the chain with the user's input.
    try:
        response = chain.run(input_word)
        output_text.delete(1.0, tk.END)
        output_text.insert(tk.END, response.strip())
    except Exception as e:
        output_text.delete(1.0, tk.END)
        output_text.insert(tk.END, f"An error occurred: {e}")

# --- Tkinter GUI Setup ---

root = tk.Tk()
root.title("LangChain Antonym Finder")
root.geometry("400x250")
root.columnconfigure(0, weight=1)

# Frame for input and button
input_frame = ttk.Frame(root, padding="10")
input_frame.grid(row=0, column=0, sticky="ew")
input_frame.columnconfigure(1, weight=1)

# Input label
input_label = ttk.Label(input_frame, text="Enter a word:")
input_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")

# Input entry box
input_entry = ttk.Entry(input_frame)
input_entry.grid(row=0, column=1, padx=5, pady=5, sticky="ew")

# Button to trigger the action
convert_button = ttk.Button(input_frame, text="Find Antonym", command=find_antonym)
convert_button.grid(row=1, column=0, columnspan=2, pady=10)

# Frame for output
output_frame = ttk.Frame(root, padding="10")
output_frame.grid(row=1, column=0, sticky="nsew")
output_frame.columnconfigure(0, weight=1)
output_frame.rowconfigure(0, weight=1)

# Output text area
output_text = tk.Text(output_frame, height=5, wrap="word")
output_text.grid(row=0, column=0, sticky="nsew")

# Start the Tkinter event loop.
root.mainloop()
```
